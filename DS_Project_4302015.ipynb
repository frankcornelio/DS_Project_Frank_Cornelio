```python
#Dowloading Data from Dataset
aapl = pd.io.data.get_data_yahoo('AAPL', start=datetime.datetime(2006, 10, 1), end=datetime.datetime(2012, 1, 1))

aapl.head()
```

```python
#Creating DataFrame
df = pd.io.data.get_data_yahoo('AAPL', start=datetime.datetime(2006, 10, 1), end=datetime.datetime(2012, 1, 1))
```

```python
df.head
```

```python
df.index
```

```python
#Calculating Difference in Open and Close Prices
df['diff'] = df.Open - df.Close
df.head()
```

```python
close_px = df['Adj Close']
```

```python
#Calculating Moving Averages
mavg = pd.rolling_mean(close_px, 40)
mavg[-10:]
```

```python
#Calculating Returns
rets = close_px / close_px.shift(1) -1
rets.head()
```

```python
#Calculating Percent Change
close_px.pct_change().head()
```

```python
#Plotting AAPL stock price and it's moving average
close_px.plot(label='AAPL')
mavg.plot(label='mavg')
plt.legend()
```

```python
#Creating DataFrame for Apple, GE, IBM, Coca Cola, Microsoft and Pepsi stock prices
df = pd.io.data.get_data_yahoo(['AAPL', 'GE', 'IBM', 'KO', 'MSFT', 'PEP'], start = datetime.datetime(2010, 1, 1), end = datetime.datetime(2013, 1, 1))['Adj Close']
df.head()
```

```python
#Percent change in returns
rets = df.pct_change()
```

```python
#Plotting returns- Pepsi vs Coca Cola
plt.scatter(rets.PEP, rets.KO)
plt.xlabel('Returns PEP')
plt.ylabel('Returns KO')
```

```python
#Plotting returns matrix
pd.scatter_matrix(rets, diagonal='kde', figsize=(10, 10));
```

```python
#Calculating Correlations
corr = rets.corr()
corr
```

```python
#Plotting returns matrix-'Heat Index'
plt.imshow(corr, cmap='hot', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns)
plt.yticks(range(len(corr)), corr.columns);
```

```python
# Plotting Expected Returns vs Risk
plt.scatter(rets.mean(), rets.std())
plt.xlabel('Expected returns')
plt.ylabel('Risk')
for label, x, y in zip(rets.columns, rets.mean(), rets.std()):
    plt.annotate(
        label, 
        xy = (x, y), xytext = (40, -40),
        textcoords = 'offset points', ha = 'right', va = 'top',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
```
